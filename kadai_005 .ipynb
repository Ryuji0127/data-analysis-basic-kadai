```python
# step1.matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する

import matplotlib.pyplot as plt
import numpy as np

# step2.6個のグラフを表示する領域を作成します。Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述する。
fig, axes = plt.subplots(2, 3, tight_layout = True) # noews = 2, ncols = 3  tight_layout = True <----サブプロットのレイアウトを自動調整するオプション

# step3.各領域に、折れ線グラフ、sin関数、ヒストグラム、散布図のデータを表示するコードを記述します。
# 折れ線グラフ axes [0,0]
y0 = [1 , 2 , -5 , 2] # 座標をを指定する
axes[0,0].plot(y0) # 描写位置指定する axes[0.0]の位置に()内数値のグラフを描写

# sin関数 axes [0,1]
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(x1, y1)

# ヒストグラム axes [0,2]
x2 = np.random .normal(15 ,5 ,2000)
axes[0,2].hist(x2)

# 散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3,y3)

# step4.グラフを表示させる行「plt.show()」を書く
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# プロット
fig, ax = plt.subplots() # 描画領域全体にサブプロットを一つ生成するコード

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

plt.show()
```
